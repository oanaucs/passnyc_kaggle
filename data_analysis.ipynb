```python
corr_coeff = school_explorer.corr(method='spearman')>0.5
corr_coeff_idxs = []

for c_1 in range(len(corr_coeff)):
    for c_2 in range(len(corr_coeff.iloc[:, c_1])):
        if corr_coeff.iloc[c_1][c_2] == True and c_1 != c_2:
            corr_coeff_idxs.append((c_1, c_2))
            
print(corr_coeff_idxs)
```

```python
# sort them by first column and then compute scatterplots
ordered_coeff = {}
for coeff_tup in corr_coeff_idxs:
    if coeff_tup[0] in ordered_coeff:
        ordered_coeff[school_explorer.columns[coeff_tup[0]]].append(school_explorer.columns[coeff_tup[1]])
    else:
        ordered_coeff.update({school_explorer.columns[coeff_tup[0]]: school_explorer.columns[[coeff_tup[1]]]})
```

```python
print(ordered_coeff)
```

```python
for c_1, columns in ordered_coeff.items():
    curr_explorer = pd.DataFrame(school_explorer, columns=columns)
    pd.plotting.scatter_matrix(curr_explorer, figsize=(4, 6))    
    
```
